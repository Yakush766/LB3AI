```python
import matplotlib.pyplot as plt

# 1. Данные для круговой диаграммы (из предыдущей задачи)
оценки = ['Отлично', 'Хорошо', 'Удовлетворительно', 'Неудовлетворительно']
количество_студентов = [20, 35, 30, 15]

# 2. Создаем круговую диаграмму
plt.figure(figsize=(8, 6)) # Настраиваем размер графика

# autopct - формат отображения процентов, startangle - угол начала отсчета
plt.pie(количество_студентов, labels=оценки, autopct='%1.1f%%', startangle=90,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'}) # Добавляем разделительные линии

# 3. Добавляем заголовок
plt.title('Распределение Оценок Студентов (Круговая Диаграмма)')

# 4. Делаем круговую диаграмму кругом (а не эллипсом)
plt.axis('equal')

# 5. Отображаем график
plt.show()
```
