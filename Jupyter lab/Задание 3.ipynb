```python
import numpy as np
import matplotlib.pyplot as plt

# 1.   
_ = 50
x = np.random.rand(_)  #  x   0  1
y = np.random.rand(_)  #  y   0  1

# 2.       x  y
 = x  #    x-
 = y * 100  #    y-,   

# 3.   
plt.figure(figsize=(8, 6))  #   

# s -  , c -  , cmap -  
plt.scatter(x, y, s=, c=, cmap='viridis', alpha=0.7)

# 4.     
plt.xlabel(' X')
plt.ylabel(' Y')
plt.title(' :   X,   Y')

# 5.   
plt.colorbar(label='  X')

# 6.  
plt.show()
```
