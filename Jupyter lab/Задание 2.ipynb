```python
import numpy as np
import matplotlib.pyplot as plt

# 1.   x
x = np.linspace(-10, 10, 400)

# 2.   y   
y1 = x  # y = x
y2 = x**2 # y = x^2
y3 = x**3 # y = x^3

# 3.  
plt.figure(figsize=(10, 8))  #      

#       
plt.plot(x, y1, 'b:', label='y = x (, )') # , 
plt.plot(x, y2, 'g-.', label='y = x^2 (, -)') # , -
plt.plot(x, y3, 'r-', label='y = x^3 (, )') # , 

# 4.     
plt.title('  y = x, y = x^2  y = x^3')
plt.xlabel('x')
plt.ylabel('y')

# 5.  
plt.grid(True)

# 6.  
plt.legend()

# 7.    
plt.axis('equal')

# 8.       (,  )
plt.xlim([-5, 5]) #     x  
plt.ylim([-5, 5]) #     y  ,   x^3

# 9.  
plt.show()
```
