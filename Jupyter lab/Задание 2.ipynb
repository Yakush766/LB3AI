```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Определяем значения x
x = np.linspace(-10, 10, 400)

# 2. Вычисляем значения y для каждой функции
y1 = x  # y = x
y2 = x**2 # y = x^2
y3 = x**3 # y = x^3

# 3. Создаем график
plt.figure(figsize=(10, 8))  # Настраиваем размер графика для лучшей видимости

# Рисуем линии с указанными цветами и стилями
plt.plot(x, y1, 'b:', label='y = x (синяя, пунктирная)') # синий, пунктир
plt.plot(x, y2, 'g-.', label='y = x^2 (зеленая, штрих-пунктирная)') # зеленый, штрих-пунктир
plt.plot(x, y3, 'r-', label='y = x^3 (красная, сплошная)') # красный, сплошной

# 4. Добавляем заголовок и подписи осей
plt.title('Графики функций y = x, y = x^2 и y = x^3')
plt.xlabel('x')
plt.ylabel('y')

# 5. Добавляем сетку
plt.grid(True)

# 6. Добавляем легенду
plt.legend()

# 7. Делаем оси одного масштаба
plt.axis('equal')

# 8. Устанавливаем пределы осей для лучшей визуализации (необязательно, но рекомендуется)
plt.xlim([-5, 5]) # Настраиваем пределы по оси x по необходимости
plt.ylim([-5, 5]) # Настраиваем пределы по оси y по необходимости, особенно для x^3

# 9. Отображаем график
plt.show()
```
